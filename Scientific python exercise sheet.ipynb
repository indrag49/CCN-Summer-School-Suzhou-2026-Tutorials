# Coding Exercises for Scientific Python and AI-Assisted Coding

## Python Basics

### 1. variables and firing rate 

A neuron fires 72 spikes during a 12-second recording. Create two variables: `num_spikes`, and `recording_time`

and compute the firing rate using:

$$
r = \frac{\text{number of spikes}}{\text{recording time}}
$$

Print the firing rate in Hz.

### 2. Lists and indexing 

Create a Python `list` containing the spike counts: `[4, 7, 1, 9, 3, 6]`

and print:

- the first spike count,
- the third spike count,
- the last spike count.

Then compute the total number of spikes across all neurons.

### 3. Loops 

Using the spike count list from exercise 2, write a for loop that computes the firing rate of each neuron.

Assume the recording duration is 8 seconds.

Store the results in a new list called:

`firing_rates`

### 4. Functions 

Write a function: `spike_count_to_rate(spike_count, recording_duration)` and return the firing rate.

Use this function to compute firing rates for at least five neurons.

### 5. Dictionaries 

Create a dictionary for one neuron containing `unit ID`, `brain area`, `recording depth`, and `mean firing rate`.

Then add one new key called: `quality` with value: `"good"`

Print all values in the dictionary.

### 6. NumPy arrays 

Create a NumPy array containing the firing rates: `[2.5, 4.1, 0.8, 7.3, 5.6, 3.2]`

Then compute:
- mean,
- maximum,
- minimum,
- standard deviation.

### 7. Vectorisation 

Assume the spike counts of six neurons are: `[12, 18, 7, 25, 10, 15]`. Use NumPy, without a loop, to compute the firing rate of each neuron for a 5 second recording.

### 8. Spike raster matrix 

Create a binary spike raster matrix with: 12 neurons and 100 time bins. Each entry should be either 0 or 1.

Print the shape of the matrix and then explain what the two dimensions represent.

### 9. Operations along NumPy axes 

Using a spike raster matrix `X`  with shape `(12, 100)` compute the following:

- total spike count for each neuron,
- population activity at each time bin,
- total number of spikes in the whole raster.

### 10. Boolean masking 

Given an array of firing rates, select only neurons whose firing rate is: greater than 4 Hz and less than 10 Hz.

Also print:

- selected firing rates,
- number of selected neurons.

### 11. Random numbers and spike generation

Generate a random binary spike raster for 30 neurons and 500 time bins

Assume each neuron has probability:

$$
p = 0.015
$$

of spiking in each bin.

Plot the raster using `matplotlib`.

### 12. Poisson spike counts

Use a Poisson distribution to simulate spike counts from 100 neurons. Use mean spike count:

$$
\lambda = 6,
$$

and plot a histogram of the spike counts. Also compute the sample mean.

### 13. Numba acceleration

Write a Python function that computes:

$$
S = \sum_{i=1}^{n} x_i^2
$$

using a loop. Then create a `Numba` accelerated version of the same function. Compare the running times using `%timeit`

### 14. Leaky neural activity model

Consider the model:

$$
\tau \frac{dx}{dt} = -x + I
$$

Simulate the model using `solve_ivp()` for:

$$
\tau = 15,  \\
I = 2,  \\
x_0 = 0.
$$

Plot $x(t)$ over the time interval: $0 \le t \le 120$

### 15. Leaky integrate-and-fire neuron

Simulate the subthreshold membrane voltage model:

$$
-(V - V_{\mathrm{rest}}) + RI,
$$

using parameter values different from the tutorial example.

- Plot the membrane voltage.
- Add a horizontal line for the threshold voltage $V_{\rm th}$

### 16. Spike threshold and reset

Extend the LIF simulation so that whenever:

$$
V(t) \geq V_{\mathrm{th}}
$$
a spike is recorded and the voltage is reset to: $V_{\rm reset}$

- Store all spike times.
- Compute the firing rate of the neuron.

### 17. Excitatory-inhibitory rate model

Simulate the Wilson-Cowan model:

$$
\frac{dE}{dt} = -E + S(w_{EE}E - w_{EI}I + P)
$$
$$
\frac{dI}{dt} = -I + S(w_{IE}E - w_{II}I + Q)
$$
Choose your own parameter values.

Plot $E(t)$ and $I(t)$ on the same figure.

### 18. Signal processing

Generate a synthetic neural signal containing two oscillatory components:

$$
\sin(2\pi f_1 t)
+
0.7\sin(2\pi f_2 t)
+
\epsilon(t)
$$

where $f_1 = 8$ Hz, $f_2 = 40$ Hz, and $\epsilon(t)$ is Gaussian noise.

Compute and plot the power spectral density using: `scipy.signal.welch`

### 19. Statistics and decoding

Simulate neural activity from two stimulus conditions. Use at least 150 trials and 25 neurons

Make a subset of neurons respond more strongly to one condition. Following that do these:

- compute the mean response of each neuron in each condition,
- perform a statistical test comparing the two conditions for one selected neuron,
- train a scikit-learn classifier to decode the stimulus condition from population activity,
- report the classification accuracy,
- plot a confusion matrix.

### 20. AI-assisted coding

Write a prompt for a standard large language model asking it to help simulate a small leaky integrate-and-fire network.

Your prompt must include:

- the mathematical equation,
