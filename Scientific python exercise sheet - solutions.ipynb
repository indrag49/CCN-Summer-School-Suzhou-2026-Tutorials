### 13. Numba acceleration

Write a Python function that computes:

$$
S = \sum_{i=1}^{n} x_i^2
$$

using a loop. Then create a `Numba` accelerated version of the same function. Compare the running times using `%timeit`

```python
import numpy as np

rng = np.random.default_rng(22)
x = rng.random(1000000)
print(x)


def sum_squares_python(x):
    total = 0.0
    for i in range(len(x)):
        total += x[i] ** 2
    return total


print("Python result:", sum_squares_python(x))

# Numba version
from numba import njit

@njit
def sum_squares_numba(x):
    total = 0.0
    for i in range(len(x)):
        total += x[i] ** 2
    return total


# First call compiles the function
print("Numba result:", sum_squares_numba(x))

%timeit sum_squares_python(x)
%timeit sum_squares_numba(x)
```

### 14. Leaky neural activity model

Consider the model:

$$
\tau \frac{dx}{dt} = -x + I
$$

Simulate the model using `solve_ivp()` for:

$$
\tau = 15,  \\
I = 2,  \\
x_0 = 0.
$$

Plot $x(t)$ over the time interval: $0 \le t \le 120$

```python
from scipy.integrate import solve_ivp


def leaky_activity(t, x, tau, I):
    dxdt = (-x[0] + I) / tau
    return [dxdt]


tau = 15
I = 2
x0 = [0]

t_eval = np.linspace(0, 120, 600)

sol = solve_ivp(
    leaky_activity,
    t_span=[0, 120],
    y0=x0,
    t_eval=t_eval,
    args=(tau, I)
)

plt.figure(figsize=(6, 4))
plt.plot(sol.t, sol.y[0])
plt.xlabel("Time")
plt.ylabel("Activity x(t)")
plt.title("Leaky neural activity model")
```

### 15. Leaky integrate-and-fire neuron

Simulate the subthreshold membrane voltage model:

$$
-(V - V_{\mathrm{rest}}) + RI,
$$

using parameter values different from the tutorial example.

- Plot the membrane voltage.
- Add a horizontal line for the threshold voltage $V_{\rm th}$

```python
def lif_subthreshold(t, V, tau_m, V_rest, R, I):
    dVdt = (-(V[0] - V_rest) + R * I) / tau_m
    return [dVdt]


tau_m = 25.0      # ms
V_rest = -70.0    # mV
V_th = -52.0      # mV
R = 12.0
I = 1.8

V0 = [-70.0]

t_eval = np.linspace(0, 150, 1000)

sol = solve_ivp(
    lif_subthreshold,
    t_span=[0, 150],
    y0=V0,
    t_eval=t_eval,
    args=(tau_m, V_rest, R, I)
)

plt.figure(figsize=(7, 4))
plt.plot(sol.t, sol.y[0], label="Membrane voltage")
plt.axhline(V_th, linestyle="--", label="Threshold")
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (mV)")
plt.title("Subthreshold LIF neuron")
plt.legend()
```

### 16. Spike threshold and reset

Extend the LIF simulation so that whenever:

$$
V(t) \geq V_{\mathrm{th}}
$$
a spike is recorded and the voltage is reset to: $V_{\rm reset}$

- Store all spike times.
- Compute the firing rate of the neuron.

```python
dt = 0.1       # ms
T = 300.0      # ms

time = np.arange(0, T, dt)

tau_m = 25.0
V_rest = -70.0
V_reset = -70.0
V_th = -52.0
R = 12.0
I = 2.2

V = np.zeros(len(time))
V[0] = V_rest

spike_times = []

for k in range(1, len(time)):
    dVdt = (-(V[k - 1] - V_rest) + R * I) / tau_m
    V[k] = V[k - 1] + dt * dVdt

    if V[k] >= V_th:
        spike_times.append(time[k])
        V[k] = V_reset

firing_rate_hz = len(spike_times) / (T / 1000)

plt.figure(figsize=(8, 4))
plt.plot(time, V)
plt.axhline(V_th, linestyle="--", label="Threshold")
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (mV)")
plt.title("LIF neuron with threshold and reset")
plt.legend()

print("Spike times:", spike_times)
print("Number of spikes:", len(spike_times))
print("Firing rate:", firing_rate_hz, "Hz")
```

### 17. Excitatory-inhibitory rate model

Simulate the Wilson-Cowan model:

$$
\frac{dE}{dt} = -E + S(w_{EE}E - w_{EI}I + P)
$$
$$
\frac{dI}{dt} = -I + S(w_{IE}E - w_{II}I + Q)
$$
Choose your own parameter values.

Plot $E(t)$ and $I(t)$ on the same figure.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def wilson_cowan(t, z, w_EE, w_EI, w_IE, w_II, P, Q):
    E, I = z

    dE = -E + sigmoid(w_EE * E - w_EI * I + P)
    dI = -I + sigmoid(w_IE * E - w_II * I + Q)

    return [dE, dI]


w_EE = 8.0
w_EI = 10.0
w_IE = 9.0
w_II = 3.0
P = 0.8
Q = 0.2

z0 = [0.1, 0.2]

t_eval = np.linspace(0, 100, 1000)

sol = solve_ivp(
    wilson_cowan,
    t_span=[0, 100],
    y0=z0,
    t_eval=t_eval,
    args=(w_EE, w_EI, w_IE, w_II, P, Q)
)

plt.figure(figsize=(7, 4))
plt.plot(sol.t, sol.y[0], label="E(t)")
plt.plot(sol.t, sol.y[1], label="I(t)")
plt.xlabel("Time")
plt.ylabel("Population activity")
plt.title("Excitatory-inhibitory rate model")
plt.legend()
```

### 18. Signal processing

Generate a synthetic neural signal containing two oscillatory components:

$$
\sin(2\pi f_1 t)
+
0.7\sin(2\pi f_2 t)
+
\epsilon(t)
$$

where $f_1 = 8$ Hz, $f_2 = 40$ Hz, and $\epsilon(t)$ is Gaussian noise.

Compute and plot the power spectral density using: `scipy.signal.welch`

```python
from scipy import signal

rng = np.random.default_rng(24)

fs = 1000
duration = 3.0

t = np.linspace(0, duration, int(fs * duration), endpoint=False)

f1 = 8
f2 = 40

epsilon = rng.normal(0, 0.5, size=len(t))

x = (
    np.sin(2 * np.pi * f1 * t)
    + 0.7 * np.sin(2 * np.pi * f2 * t)
    + epsilon
)

plt.figure(figsize=(8, 4))
plt.plot(t, x)
plt.xlabel("Time (s)")
plt.ylabel("Signal")
plt.title("Synthetic neural signal")

freqs, power = signal.welch(x, fs=fs)

plt.figure(figsize=(6, 4))
plt.semilogy(freqs, power)
plt.xlim(0, 80)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power")
plt.title("Power spectral density")
```

### 19. Statistics and decoding

Simulate neural activity from two stimulus conditions. Use at least 150 trials and 25 neurons

Make a subset of neurons respond more strongly to one condition. Following that do these:

- compute the mean response of each neuron in each condition,
- perform a statistical test comparing the two conditions for one selected neuron,
- train a scikit-learn classifier to decode the stimulus condition from population activity,
- report the classification accuracy,
- plot a confusion matrix.

```python
from scipy import stats

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

rng = np.random.default_rng(24)

n_trials = 162
n_neurons = 28

# Two stimulus conditions
y = np.repeat([0, 1], n_trials // 2)

# Baseline neural activity
X = rng.normal(5, 1, size=(n_trials, n_neurons))

# Stimulus condition 1 increases activity in first 6 neurons
X[y == 1, :6] += 1.2

# Mean response of each neuron in each condition
mean_condition_0 = X[y == 0].mean(axis=0)
mean_condition_1 = X[y == 1].mean(axis=0)

print("Mean response, condition 0:")
print(mean_condition_0)

print("\nMean response, condition 1:")
print(mean_condition_1)

# Statistical test for one neuron
neuron_index = 0

test_result = stats.ttest_ind(
    X[y == 0, neuron_index],
    X[y == 1, neuron_index]
)

print("\nSelected neuron:", neuron_index)
print("t statistic:", test_result.statistic)
print("p-value:", test_result.pvalue)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

classifier = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000)
)

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("\nClassification accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(4, 4))
plt.imshow(cm, cmap="Blues")
plt.xlabel("Predicted condition")
plt.ylabel("True condition")
plt.title("Confusion matrix")
plt.colorbar()
```
