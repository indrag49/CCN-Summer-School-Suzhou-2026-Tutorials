```python
## First example

## A normal Python example

def sum_squares_python(x):
    total = 0.0

    for i in range(len(x)):
        total += x[i] ** 2

    return total
```

```python
## A numba version

from numba import njit

@njit # use the decorator
def sum_squares_numba(x):
    total = 0.0

    for i in range(len(x)):
        total += x[i] ** 2

    return total
```

```python
## Use it

x = np.random.random(1000000)

print(sum_squares_python(x))
print(sum_squares_numba(x))
```

```python
%timeit sum_squares_python(x)
%timeit sum_squares_numba(x)
```

## A neuroscience example

### Suppose each row is a feature vector for a neuron, and we want pairwise Euclidean distances.

```python
# -------------------------------
# 1. Simple Python version
# -------------------------------
def pairwise_distances_python(X):
    n, d = X.shape
    D = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            total = 0.0

            for k in range(d):
                diff = X[i, k] - X[j, k]
                total += diff * diff

            D[i, j] = np.sqrt(total)

    return D
```

```python
# -------------------------------
# 2. Numba version
# -------------------------------
@njit
def pairwise_distances_numba(X):
    n, d = X.shape
    D = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            total = 0.0

            for k in range(d):
                diff = X[i, k] - X[j, k]
                total += diff * diff

            D[i, j] = np.sqrt(total)

    return D
```

```python
# -------------------------------
# 3. Create example data
# -------------------------------
rng = np.random.default_rng(42)

n = 300      # number of points / neurons
d = 50       # number of features / time bins

X = rng.random((n, d))


# -------------------------------
# 4. Warm up Numba
# -------------------------------
# The first call compiles the function.
# Do not include this compilation time in the timing comparison.
D_numba = pairwise_distances_numba(X)


# -------------------------------
# 5. Check correctness
# -------------------------------
D_python = pairwise_distances_python(X)

print("Are results close?", np.allclose(D_python, D_numba))
```

```python
# First call compiles
pairwise_distances_numba(X)

# Then time it
%timeit pairwise_distances_python(X)
%timeit pairwise_distances_numba(X)
```

## parallel `numba` with `prange`

```python
from numba import njit, prange

## Use prange when different loop iterations are independent.

@njit(parallel=True)
def pairwise_distances_parallel(X):
    n, d = X.shape
    D = np.zeros((n, n))

    for i in prange(n):
        for j in range(n):
            total = 0.0

            for k in range(d):
                diff = X[i, k] - X[j, k]
                total += diff * diff

            D[i, j] = np.sqrt(total)

    return D
```

```python
D_parallel = pairwise_distances_parallel(X)
D_parallel
```

```python
%timeit pairwise_distances_python(X)
%timeit pairwise_distances_numba(X)
%timeit pairwise_distances_parallel(X)
```
